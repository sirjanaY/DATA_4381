```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("processed_panel_2021.csv")
print(df.shape)
print(df.dtypes)
```

```python
df["MONTH"] = pd.to_numeric(df["MONTH"], errors="coerce")

weight_col = "LNKFW1MWT"
use_weights = weight_col in df.columns

def wmean(x, w):
    return np.average(x, weights=w) if w is not None else x.mean()

print(df.head())
print(df.columns.tolist())
```

```python
df["wage_group"] = np.where(df["LowWage"] == 1, "Low-Wage", "Other-Wage")

if use_weights:
    group_counts = df.groupby("wage_group")[weight_col].sum()
else:
    group_counts = df["wage_group"].value_counts()

share = (group_counts / group_counts.sum()).sort_index()

fig, ax = plt.subplots()
share.plot(kind="bar", ax=ax)
ax.set_ylabel("Share of workers")
ax.set_title("Wage-Group Composition (CPS 2021 Panel)")
plt.tight_layout()
plt.show()

print("Counts by wage group:")
print(group_counts)
print("\nShare by wage group:")
print(share)
```

```python
if use_weights:
    trend = (
        df.groupby(["MONTH", "wage_group"])
          .apply(lambda g: wmean(g["found_job"], g[weight_col]))
          .reset_index(name="job_finding_rate")
    )
else:
    trend = (
        df.groupby(["MONTH", "wage_group"])["found_job"]
          .mean()
          .reset_index(name="job_finding_rate")
    )

fig, ax = plt.subplots()
for grp, sub in trend.groupby("wage_group"):
    ax.plot(sub["MONTH"], sub["job_finding_rate"], marker="o", label=grp)

ax.axvline(x=7, linestyle="--", alpha=0.7, label="UI termination (July)")
ax.set_xlabel("Month (2021)")
ax.set_ylabel("Job-finding probability")
ax.set_title("Job-Finding Trends by Wage Group")
ax.legend()
plt.tight_layout()
plt.show()

trend.head()
```

```python
states = df[["STATEFIP", "TreatState"]].drop_duplicates()

state_counts = states["TreatState"].value_counts().rename(
    index={0: "Control", 1: "Treated"}
)

fig, ax = plt.subplots()
state_counts.plot(kind="bar", ax=ax)
ax.set_ylabel("Number of states")
ax.set_title("Early UI Termination vs Non-Termination States")
plt.tight_layout()
plt.show()

print(states.head())
print("\nState counts:")
print(state_counts)
```

```python


weight_col = "WTFINL" if "WTFINL" in df.columns else None
df["_w"] = df[weight_col] if weight_col else 1.0

def wmean(s, w):
    s, w = s.astype(float), w.astype(float)
    return (s * w).sum() / w.sum()

rows = []
for lw, g in df.groupby("LowWage"):
    w = g["_w"]
    rows.append({
        "Group": "Low Wage" if lw == 1 else "High Wage",
        "Mean found_job": wmean(g["found_job"], w),
        "Avg Age": wmean(g["AGE"], w),
        "Share Treated States": wmean(g["TreatState"], w),     
        "COVID Avg (log cases)": wmean(g["log_monthly_cases"], w),
    })

summary = pd.DataFrame(rows).set_index("Group")

summary_fmt = summary.copy()
summary_fmt["Mean found_job"] = summary["Mean found_job"].round(3)
summary_fmt["Avg Age"] = summary["Avg Age"].round(1)
summary_fmt["Share Treated States"] = (summary["Share Treated States"]*100).round(1).astype(str) + "%"
summary_fmt["COVID Avg (log cases)"] = summary["COVID Avg (log cases)"].round(2)

summary_fmt
```
